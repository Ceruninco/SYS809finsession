```python
from tensorflow.keras.applications import VGG16

conv_base = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(224, 224, 3))

```

```python
import os
import cv2
import numpy as np

def load_data (data_path,img_shape,prefix):

    img_rows=img_shape[0]
    img_cols=img_shape[1]
    num_channel=img_shape[2]

    labels=[]
    data=[]
    len_list_img=0
    num_class=0

    img_list=os.listdir(data_path)    
    for img in img_list:
        input_img=cv2.imread(data_path + '/'+ img )
        input_img=cv2.resize(input_img,(img_rows,img_cols))
        input_img = tf.keras.applications.vgg16.preprocess_input(input_img, data_format=None)
        data.append(input_img)
        index = int(img.replace(prefix, '').replace('-', '').replace('.jpg', ''))
        labels.append(index)
           
    data = np.array(data)
    data = data.astype('float32')
    data /= 255
    labels=np.array(labels)
    print('data shape',data.shape)
    print('labels shape',labels.shape)
    return data, labels
```
